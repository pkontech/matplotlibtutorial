# Stacked Bar Graph

# Pie Chart

```python
data = df.loc[df["Year"] == 2021]
labels = list(data.columns[1:5])
slices = data.iloc[0][df.columns[1:5]]

plt.pie(slices, labels=labels, autopct='%1.1f%%')
plt.title("Uber Revenue mix for 2021")
plt.legend()
plt.show()
```

# The Matplotlib Object Hierarchy


```python
plt.figure(figsize=(18,6))
ax1 = plt.subplot(131)
ax1.bar(df["Year"], df["Mobility"])
ax1.set_title("Mobility")

ax2 = plt.subplot(132)
ax2.bar(df["Year"], df["Delivery"])
ax2.set_title("Delivery")

ax3 = plt.subplot(133)
ax3.bar(df["Year"], df["Freight"])
ax3.set_title("Freight")

plt.show()

```
